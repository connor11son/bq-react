### Step 0: Installation


*   Install `bqapi` using `pip install bqapi-ucsb`
*   Install `matplotlib` using `pip install matplotlib`



### Step 1: Import necessary packages

```python
from bqapi.comm import BQSession
from matplotlib import pyplot as plt
```

### Step 2: Authenticate the user and initiate a session

```python
bq = BQSession().init_local("Username", "Password", 
                            bisque_root = "https://bisque2.ece.ucsb.edu")
```

### Step 3: Get the file URI and Download the file

URI can be found at the end of the URL.

**Sample URL**: [https://bisque2.ece.ucsb.edu/.../data_service/00-3JemvVz28JrHJJYQQq8K3o](https://bisque2.ece.ucsb.edu/client_service/view?resource=https://bisque2.ece.ucsb.edu/data_service/00-3JemvVz28JrHJJYQQq8K3o)

Hence the **URI** is going to be `00-3JemvVz28JrHJJYQQq8K3o`






```python
### Get the URI of the Image 
predictor_url = bq.service_url('image_service', path='00-3JemvVz28JrHJJYQQq8K3o')

### Download the Image using `fetchblob`
bq.fetchblob(predictor_url, 'sample_image.jpeg')
```

### Step 4 (Optional): View the downloaded file
Use approporiate packages to make sure the download was succesful.

```python
plt.imshow(plt.imread('sample_image.jpeg'), cmap="gray")
```
